```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j])> threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return list(col_corr)
```

```python
col_rm = correlation(df, 0.7)
```

```python
col_rm
```

```python
data = df.loc[:, list(filter(lambda x : x not in col_rm, df.columns))]
```

```python
def has_zero(dataf):
  col_rm = set()

  for i in range(len(dataf.columns)):
    colname = dataf.columns[i]
    if len(dataf[colname].unique()) == 1:
      col_rm.add(colname)
  return col_rm
```

```python
col_zero = has_zero(data)
```

```python
data = data.loc[:, list(filter(lambda x : x not in col_zero, data.columns))]
```
